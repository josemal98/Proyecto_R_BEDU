## 1.1 Descripción del proyecto

Este proyecto forma parte del último módulo "Programación y estadística con R" del curso de Ciencia de Datos de BEDU y se centra en identificar los determinantes clave que influyen en la esperanza de vida. La pregunta central que buscamos responder es esencial para las organizaciones de salud que buscan mejorar la esperanza de vida en regiones específicas: ¿Qué variables, y de qué manera, pueden modificarse para mejorar eficazmente la esperanza de vida?

El proyecto parte de un conjunto de datos conformado por valores de esperanza de vida, asociados a factores económicos y de salud. Se busca, por medio de un análisis exhaustivo, identificar y cuantificar el impacto de las diversas variables independientes en la variable dependiente: la esperanza de vida. ## 1.2 Descripción del Dataset

Este dataset se generó fusionando información de dos fuentes distintas. La primera fuente proviene del Observatorio Mundial de la Salud (GHO), parte de la Organización Mundial de la Salud (OMS), que realiza un seguimiento de la salud y otros factores relacionados en 193 países. Esta información aborda aspectos de salud, incluyendo la esperanza de vida y otros indicadores relevantes. La segunda fuente corresponde al sitio web de las Naciones Unidas, proporcionando datos económicos correspondientes a los mismos países.

Se seleccionaron los factores críticos más representativos entre todas las categorías de aspectos relacionados con la salud. Se sabe que ha habido un notable desarrollo en el sector de la salud a lo largo del presente siglo, mejorando las tasas de mortalidad, especialmente en países en desarrollo, en comparación con las últimas tres décadas. Por ende, el dataset abarca datos desde el año 2000 hasta el 2015 para los 193 países, permitiendo un análisis exhaustivo.

Una inspección visual inicial reveló algunos valores faltantes, principalmente en variables como la población, Hepatitis B y el Producto Interno Bruto (PIB). Dada la significativa proporción de datos faltantes, aproximadamente el 50%, la opción más viable es aplicar imputación. Como se explicó en la introducción, se ha observado que la falta de datos se concentra principalmente en países menos conocidos. Esto sugiere que la ausencia de estos datos podría deberse a su no disponibilidad o incluso a su inexistencia. Por lo tanto, imputar simplemente asumiendo que estos valores correspondían a ceros resulta inviable.

En cambio, una alternativa más adecuada es emplear algún valor de tendencia central. Esto reduce nuestra elección a decidir entre la media y la mediana para cada variable. Tomando en cuenta la menor sensibilidad de la mediana frente a la media con respecto a los valores atípicos, se opta por tomar esta decisión de manera individual para cada variable, considerando la cantidad de este tipo de valores en cada una. Mientras que en aquellas variables que presentan valores atípicos se elige la mediana, en aquellas sin este tipo de valores se utiliza la media. Para identificar estos valores atípicos en cada variable se utiliza el criterio de 1.5 veces el rango intercuartílico.

```r
# Crear un data frame para la gráfica
missing_data <- data.frame(
  variable = names(df_rename),
  datos_faltantes = colSums(is.na(df_rename))
)

# Seleccionar las variables con datos faltantes
variables_con_faltantes <- missing_data$variable[missing_data$datos_faltantes > 0]

# Función para contar outliers en una variable con base en el criterio 1.5 * IQR
count_outliers <- function(variable) {

  # Eliminar NA's
  variable <- variable[!is.na(variable)]

  # Contar outliers solo si hay valores numéricos
  if (length(variable) > 0) {
    # Calcular el rango intercuartílico (IQR)
    iqr <- IQR(variable)

    # Calcular los límites para identificar outliers
    lower_limit <- quantile(variable, 0.25) - 1.5 * iqr
    upper_limit <- quantile(variable, 0.75) + 1.5 * iqr

    # Contar outliers
    num_outliers <- sum(variable < lower_limit | variable > upper_limit)

    return(num_outliers)
  } else {
    return(NA)
  }
}

# Contar outliers solo para las variables con datos faltantes
outliers_count_per_variable <- sapply(df_rename[variables_con_faltantes], count_outliers)

# Crear un data frame con la cantidad de outliers por variable
outliers_count_df <- data.frame(
  variable = names(outliers_count_per_variable),
  outliers_count = outliers_count_per_variable
)

# Ordenar el data frame por la cantidad de outliers de mayor a menor
outliers_count_df <- outliers_count_df[order(-outliers_count_df$outliers_count), ]

# Resetting index numbers of rows
rownames(outliers_count_df) <- NULL

cat("Cantidad de outliers en variables con valores faltantes:\n\n\n")

# Imprimir los resultados ordenados
outliers_count_df
```

```r
# Crear un data frame para la gráfica
missing_data <- data.frame(
  variable = names(df_rename),
  datos_faltantes = colSums(is.na(df_rename))
)

# Seleccionar las variables con datos faltantes
variables_con_faltantes <- missing_data$variable[missing_data$datos_faltantes > 0]

# Crear la figura con 3 renglones y 5 columnas
par(mfrow = c(3, 5), mar = c(4, 4, 2, 1))

# Iterar sobre las variables y crear un boxplot para cada una
for (variable in variables_con_faltantes) {

  # Obtener el color según outliers_count_df
  color <- ifelse(outliers_count_df[outliers_count_df$variable == variable, "outliers_count"] == 0, "lightblue", "lightcoral")

  # Crear el boxplot con el color correspondiente
  boxplot(df_rename[[variable]], main = variable, col = color, border = "black", notch = TRUE)
}

# Restablecer el diseño de la figura a uno por defecto
par(mfrow = c(1, 1), mar = c(5, 4, 4, 2) + 0.1)
```

#### Imputación de datos faltantes

Se imputa la media a los valores faltantes de las variables que no tienen outliers (boxplots azules), en tanto que se usa la mediana para aquellas variables que sí tienen outliers (boxplots rojos).

```r
# Crear un vector para almacenar los valores de imputación
imputacion <- c()

# Iterar sobre las filas de outliers_count_df
for (i in 1:dim(outliers_count_df)[1]) {
  # Obtener el nombre de la variable actual
  variable <- outliers_count_df$variable[i]

  # Verificar si hay outliers en la variable actual
  if (outliers_count_df$outliers_count[i] > 0) {
    # Imputar con la mediana si hay outliers
    imputacion[i] <- median(df_rename[, variable], na.rm = TRUE)
  } else {
    # Imputar con la media si no hay outliers
    imputacion[i] <- mean(df_rename[, variable], na.rm = TRUE)
  }
}

# Aplicar la imputación al dataframe original (df_rename)
for (i in 1:dim(outliers_count_df)[1]) {
  # Obtener el nombre de la variable actual
  variable <- outliers_count_df$variable[i]

  # Imputar los valores faltantes en df con los valores calculados
  df_rename[, variable][is.na(df_rename[, variable])] <- imputacion[i]
}
```

## 2.3 Datos atípicos

Aunque previamente aplicamos el criterio de 1.5 veces el rango intercuartílico para identificar valores atípicos en cada variable por separado, eliminar todos los datos considerados outliers según este criterio sería una medida extrema. Por ende, optamos por un enfoque que permitiera identificar valores atípicos considerando simultáneamente todas las variables del dataframe. En este sentido, hemos elegido utilizar el método de Local Outlier Factor o LOF, que está disponible como función de la librería "dbscan". Este enfoque ofrece una perspectiva más holística al abordar la detección de outliers, considerando la interacción de las variables en conjunto.

```r
# Generar un ejemplo de dataframe (reemplaza esto con tu dataframe real)
set.seed(123)

# Crear un dataframe solo con las columnas numéricas
X <- df_rename[, !colnames(df_rename) %in% "Status"]

# Calcular el factor de atípicidad local (Local Outlier Factor)
lof_resultados <- lof(X, minPts = 2)

# Obtener la cantidad de filas identificadas como outliers
cantidad_outliers <- sum(lof_resultados > 3)

# Obtener el porcentaje de outliers respecto al total de filas
porcentaje_outliers <- (cantidad_outliers / nrow(X)) * 100

# Imprimir resultados
cat("Cantidad de filas identificadas como outliers:", cantidad_outliers, "\n")
cat("Porcentaje de outliers respecto al total de filas:", porcentaje_outliers, "%\n")
```

```r
# Descartamos los outliers de nuestro DataFrame
df_clean <- df_rename[lof_resultados <= 3, ]
```

```r
# Dimensiones del nuevo DataFrame
cat("Dataset size: (",toString(dim(df_clean)),")",sep="")
```

Re-evaluamos la cantidad de outliers de cada variable utilizando el critero del 1.5 veces el rango intercuartílico. Comparamos esta cantidad de outliers, con la cantidad obtenida antes de realizar la eliminación de datos atípicos con Local Outlier Factor.

```r
library(tidyr)

# Función para calcular outliers_count con el criterio de 1.5 veces el rango intercuartílico
calculate_outliers_count <- function(data) {
  outliers_count_df <- data %>%
    summarise(across(everything(), ~ sum(boxplot.stats(.)$out > 1.5 * IQR(.)))) %>%
    pivot_longer(cols = everything(), names_to = "variable", values_to = "outliers_count") %>%
    filter(outliers_count > 0)
  return(outliers_count_df)
}

# Calcular outliers_count_df_rename
outliers_count_df_rename <- calculate_outliers_count(df_rename)

# Calcular outliers_count_df_clean
outliers_count_df_clean <- calculate_outliers_count(df_clean)

# Filtrar variables numéricas con outliers
numerical_columns_rename <- outliers_count_df_rename$variable
numerical_columns_clean <- outliers_count_df_clean$variable

# Filtrar las variables en outliers_count_df_rename y outliers_count_df_clean que coincidan con las numéricas
outliers_count_rename <- outliers_count_df_rename %>%
  filter(variable %in% numerical_columns_rename)

outliers_count_clean <- outliers_count_df_clean %>%
  filter(variable %in% numerical_columns_clean)

# Combinar los dos dataframes
combined_df <- bind_rows(
  mutate(outliers_count_rename, Dataset = "Antes de LOF"),
  mutate(outliers_count_clean, Dataset = "Después de LOF")
)

# Ajustar el tamaño de los puntos
point_size <- 6

# Graficar scatter plot de la cantidad de outliers antes y después
ggplot(combined_df, aes(x = reorder(variable, -outliers_count), y = outliers_count, color = Dataset)) +
  geom_point(size = point_size) +
  labs(title = "Cantidad de outliers antes y después de LOF",
       x = "Variable",
       y = "Cantidad de Outliers") +
  scale_color_manual(values = c("Antes de LOF" = "lightcoral", "Después de LOF" = "lightblue")) +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 18, hjust = 0.5, face = "bold"),
    axis.title = element_text(size = 14),
    axis.text = element_text(size = 12),
    axis.text.x = element_text(angle = 45, hjust = 1),  # Rotar nombres del eje x
    legend.position = "top",
    legend.title = element_text(size = 14),
    legend.text = element_text(size = 12),
    #panel.grid = element_blank(),  # Eliminar el grid del fondo
    axis.ticks.y = element_blank()  # Eliminar las marcas de los ticks en el eje y
  )
```

Notamos que la mayor reducción de datos atípicos ocurre en las variables que originalmente presentaban mayor cantidad de estos.

## 2.4 Transformación de los datos

A continuación, iniciamos un análisis examinando la distribución de la variable dependiente. Identificamos cualquier tendencia o sesgo que pueda presentar y que pueda influir en análisis subsiguientes. Basándonos en estas observaciones, evaluamos métodos de escalamiento que mitiguen los efectos identificados.

En cuanto a las variables independientes, aplicamos un enfoque consistente de estandarización para todas ellas. Este proceso asegura que las diferentes magnitudes de las variables no afecten los ajustes de modelos y análisis subsiguientes.

Sin embargo, una alta correlación sigue sin necesariamente indicar colinealidad. Por lo tanto, a continuación complementamos el análisis con la métrica del Factor de Inflación de la Varianza (FIV). Esta herramienta permite cuantificar la intensidad de la multicolinealidad en un análisis de regresión normal de mínimos cuadrados.

Aunque no existe un criterio estricto de a partir de qué valores de FIV se considera que la colinealidad entre las variables es alta, por lo general se siguen las siguientes pautas:

- **VIF < 5:** colinealidad baja.
- **5 < VIF < 10:** colinealidad moderada.
- **VIF < 10:** colinealidad alta.

Con base en esto, decidimos considerar como problemático a aquellas variables de VIF por encima de 5.

```r
mod.linear <- lm(Life_expectancy~ ., data = subset(df_trans, select =-c(Status)))

# Calcular el VIF
vif_values <- car::vif(mod.linear)

# Crear un dataframe con los valores de VIF y las características correspondientes
vif_data <- data.frame(Variables = names(vif_values), VIF = vif_values)

# Ordenar el dataframe por VIF de manera descendente
vif_data <- vif_data[order(-vif_data$VIF), ]

# Definir colores basados en el umbral de VIF
vif_data$Color <- ifelse(vif_data$VIF > 5, "lightcoral", "lightblue")

# Crear la gráfica de barras con orden descendente
ggplot(vif_data, aes(x = reorder(Variables, -VIF), y = VIF, fill = Color)) +
  geom_bar(stat = "identity", color = "white") +  # Eliminar bordes

  # Etiquetas de datos (valores de las barras) encima de cada barra
  geom_text(aes(label = round(VIF, 2), y = VIF + 3), size = 4, color = "black") +

  # Línea horizontal en el valor 5 del eje y
  geom_hline(yintercept = 5, linetype = "dashed", color = "red") +

  # Tema y colores
  theme_minimal() +
  theme(
    panel.grid.major = element_blank(),  # Eliminar las líneas de la cuadrícula principal
    panel.grid.minor = element_blank(),  # Eliminar las líneas de la cuadrícula secundaria
    panel.background = element_rect(fill = "white", color = "white"),  # Fondo blanco sin borde
    axis.text.x = element_text(angle = 45, hjust = 1, size = 12),  # Ángulo del texto del eje x
    axis.text.y = element_text(size = 12),  # Tamaño del texto del eje y
    axis.title.x = element_text(size = 14),  # Tamaño del título del eje x
    axis.title.y = element_text(size = 14),  # Tamaño del título del eje y
    plot.title = element_text(size = 16, margin = margin(b = 20)),  # Tamaño del título
    legend.title = element_blank(),  # Eliminar el título de la leyenda
    legend.text = element_text(size = 12),  # Tamaño del texto de la leyenda
    legend.position = "top"  # Posición de la leyenda
  ) +

  # Etiquetas y título
  labs(
    x = "Características",
    y = "VIF",
    title = "VIF de cada característica"
  ) +

  # Leyenda personalizada
  scale_fill_manual(values = c("lightcoral" = "lightcoral", "lightblue" = "lightblue"),
                    labels = c("VIF <= 5", "VIF > 5"))
```

## **4. Modelado**

El propósito de esta sección es llevar a cabo la selección, entrenamiento y validación de un modelo predictivo de regresión múlitple. El modelo se nutre con los valores de las características independientes de nuestros datos, junto con sus correspondientes valores asociados de la variable dependiente. El objetivo es capacitar al modelo para que pueda predecir con precisión los valores de esta última utilizando únicamente un conjunto específico de características. Por consiguiente, la primera fase de esta sección se centra en la cuidadosa elección del conjunto de características más relevante para la variable dependiente.

## 4.1 Selección de características

En primer lugar, antes de siquiera pensar en categorizar a las características como más o menos relevantes, es necesario establecer la o las metodologías a utilizar para realizar esta selección. Decidimos probar los siguientes algoritmos de elección de características:

-   **Mejor subconjunto:** consiste en probar todas las posibles combinaciones de las variables predictoras y luego seleccionar el mejor subconjunto.

-   **Inclusión hacia adelante:** los predictores se agregan uno a la vez, comenzando con el predictor que tiene la correlación más alta con la variable dependiente. Se ingresan primero las variables de mayor importancia teórica. Una vez en la ecuación, la variable permanece allí.

-   **Eliminación hacia atrás:** todas las variables independientes se ingresan primero en la ecuación y cada una se elimina una a la vez si no contribuyen a la ecuación de regresión.

A su vez, es necesario establecer la métrica que utilizará cada uno de estos algoritmos para considerar como mejor o peor al conjunto de características seleccionadas. Nosotros decidimos probar con las siguientes métricas:

-   **Suma residual de cuadrados (rss)**
-   **R^2 ajustado (adjr2)**
-   **Cp de Mallows (cp)**
-   **Criterio de información bayesiano (bic)**

Como primer paso, probaremos el método de **Mejor subconjunto** con cada una de las métricas. A partir de esto, seleccionaremos aquella metrica con menor cantidad de características.

```r
set_plot_dimensions(16,12)

# Mejor subconjunto
regfit.best <- regsubsets(Life_expectancy~., data= df_EDA, nvmax = 16)  # Ajuste de regresión para el mejor subconjunto de variables
reg.summary <- summary(regfit.best)

par(mfrow=c(2,2))  # Configurar diseño de gráficos en una cuadrícula 2x2

# Gráfico de Suma Residual de Cuadrados (RSS)
plot(reg.summary$rss, xlab="Número de Variables", ylab="RSS", type="l")
cat("Número óptimo de varaibles RSS:", which.min(reg.summary$rss))  # Número óptimo de variables según el mínimo RSS
points(16, reg.summary$rss[16], col="red", cex=2, pch=20)  # Marcar el mínimo RSS con un punto rojo

# Gráfico de R^2 ajustado con su valor más alto
plot(reg.summary$adjr2, xlab="Número de Variables", ylab="Adjusted Rsq", type="l")
cat("\nNúmero óptimo de varaibles adjr2:", which.max(reg.summary$adjr2))  # Número óptimo de variables según el máximo R^2 ajustado
points(15, reg.summary$adjr2[15], col="red", cex=2, pch=20)  # Marcar el máximo R^2 ajustado con un punto rojo

# Gráfico de Mallow's Cp con su valor más bajo
plot(reg.summary$cp, xlab="Número de Variables", ylab="Cp", type='l')
cat("\nNúmero óptimo de varaibles CP:", which.min(reg.summary$cp))  # Número óptimo de variables según el mínimo Cp
points(13, reg.summary$cp[13], col="red", cex=2, pch=20)  # Marcar el mínimo Cp con un punto rojo

# Gráfico de BIC con su valor más bajo
plot(reg.summary$bic, xlab="Número de Variables", ylab="BIC", type='l')
cat("\nNúmero óptimo de varaibles BIC:", which.min(reg.summary$bic))  # Número óptimo de variables según el mínimo BIC
points(11, reg.summary$bic[11], col="red", cex=2, pch=20)  # Marcar el mínimo BIC con un punto rojo
```

La cantidad de variables sugerida por cada métrica fue:

*   **RSS:** 16 variables
*   **adjr2:** 15 variables
*   **CP:** 13 variables
*   **BIC:** 11 variables

Con base en esto, optamos por utilizar BIC como métrica para elegir subconjuntos.

Procedomos a evaluar el mejor subonjunto de características por medio de los otros algoritmos de seleccion, utilizando BIC como métrica.

**Inclusión hacia adelante**

```r
#Inlcusión hacia adelante.
par(mfrow=c(1,1))

regfit.fwd <- regsubsets(Life_expectancy~.,data=df_EDA,nvmax=16,method="forward")
fwd.summary <-summary(regfit.fwd)

# fwd.summary
set_plot_dimensions(8,6)
plot(fwd.summary$bic,xlab="Número de Variables",ylab="BIC",type='l')
cat("\nNúmero óptimo de variables:", which.min(fwd.summary$bic))
points(11,fwd.summary$bic[11],col="red",cex=2,pch=20)
```

La inclusión hacia adelante con la métrica BIC también sugirió 11 variables como el mejor subconjunto.

**Eliminación hacia atrás**

```r
#Eliminación hacia atrás.
regfit.bwd <- regsubsets(Life_expectancy~.,data=df_EDA,nvmax=16,method="backward")
bwd.summary <- summary(regfit.bwd)

# fwd.summary
set_plot_dimensions(8,6)
plot(bwd.summary$bic,xlab="Número de Variables",ylab="BIC",type='l')
cat("\nNúmero óptimo de variables:", which.min(bwd.summary$bic))
points(11,bwd.summary$bic[11],col="red",cex=2,pch=20)
```

La eliminación hacia atrás con la métrica BIC también sugirió 11 variables como el mejor subconjunto.

Comprobemos cuáles son las variables de cada método y cuáles son sus respectivos coeficientes en el modelo de regresión.

```r
# Instalar y cargar el paquete cowplot si no está instalado
if (!requireNamespace("cowplot", quietly = TRUE)) {
  install.packages("cowplot")
}
library(cowplot)

# Obtener los nombres de las variables del mejor subconjunto
v_names <- rownames(as.data.frame(coef(regfit.best, 12)))

# Crear un dataframe para almacenar coeficientes y valores correspondientes
coefs <- data.frame(v_names)
coefs$best_coef_value <- coef(regfit.best, 12)
coefs$fwd_coef_value <- coef(regfit.fwd, 12)
coefs$bwd_coef_value <- coef(regfit.bwd, 12)

# Configurar dimensiones de la gráfica
set_plot_dimensions(18, 14)  # Ajustar la altura para acomodar tres gráficos

# Gráfico para el mejor subconjunto
plot1 <- ggplot(coefs, aes(x = v_names, y = best_coef_value, fill = best_coef_value)) +
  geom_bar(stat = "identity") +
  ggtitle("Método Mejor subconjunto") +
  xlab("") +
  ylab("Coeficiente") +
  theme(axis.text.x = element_text(angle = 20, hjust = 1)) +
  theme(text = element_text(size = 18))

# Gráfico para la inclusión hacia adelante
plot2 <- ggplot(coefs, aes(x = v_names, y = fwd_coef_value, fill = fwd_coef_value)) +
  geom_bar(stat = "identity") +
  ggtitle("Método Inclusión hacia adelante") +
  xlab("") +
  ylab("Coeficiente") +
  theme(axis.text.x = element_text(angle = 20, hjust = 1)) +
  theme(text = element_text(size = 18))

# Gráfico para la eliminación hacia atrás
plot3 <- ggplot(coefs, aes(x = v_names, y = bwd_coef_value, fill = bwd_coef_value)) +
  geom_bar(stat = "identity") +
  ggtitle("Método Eliminación hacia atrás") +
  xlab("Características") + ylab("Coeficiente") +
  theme(axis.text.x = element_text(angle = 20, hjust = 1)) +
  theme(text = element_text(size = 18))

# Añadir un título superior a la figura
combined_plots <- plot_grid(plot1, plot2, plot3, ncol = 1, align = "v", axis = "l") +
  plot_annotation(
    title = "Comparación de métodos de selección de características",
    theme = theme(plot.title = element_text(size = 25, hjust = 0.5, vjust=0.5))
  )

# Imprimir la figura
print(combined_plots)
```

## 4.2 Entrenamiento y validación

En esta sección, analizaremos los resultados de entrenar un modelo de regresión lineal múltiple con 3 conjuntos de datos distintos:

- **df_FULL:** conjunto con las 19 variables independientes iniciales.

- **df_EDA:** conjunto con las 16 variables que restaron tras la eliminación de aquellas con alta colinealidad.

- **df_FS:** conjunto con las 11 variables elegidas por los algoritmos de selección.

Esto permitirá identificar si las variables presentes en un conjunto pero no en los otros, en verdad tienen un aporte en el poder predictivo del modelo.

#### **Modelo de regresión lineal múltiple df_FULL.**

```r
# Dimensiones de las gráficas
set_plot_dimensions(16,10)
```

```r
# Establecer semilla para reproducibilidad
set.seed(123)

# Crear una muestra aleatoria para dividir los datos en conjunto de entrenamiento y prueba
sample <- sample(c(TRUE, FALSE), nrow(df_FULL), replace=TRUE, prob=c(0.70, 0.30))

# Seleccionar datos de entrenamiento usando la muestra aleatoria
train <- df_FULL[sample, ]

# Seleccionar datos de prueba excluyendo los datos de entrenamiento
x.test <- df_FULL[!sample, ]

# Extraer la variable dependiente (Life_expectancy) de los datos de prueba
y.test <- df_FULL[!sample, ]$Life_expectancy

# Ajustar un modelo de regresión lineal utilizando todos los predictores (variables independientes) en el conjunto de entrenamiento
model.full <- lm(Life_expectancy ~ ., data = train)

# Resumen del modelo lineal ajustado
summary(model.full)
```

Podemos notar que **no** todas las variables son significativas. En particular, nos damos cuenta que entre estas variables hay algunas que fueron eliminadas en pasos posteriores. Esto podría sugerir que realidad la información que aportan es redundante al modelo.

Comprobamos las mértricas RMSE, $R^2$ ajustado y los supuestos de linealidad para model.full.




```r
# Realizar predicciones con el modelo completo en el conjunto de prueba
pred <- predict(model.full, newdata=x.test)

# Calcular y mostrar el RMSE entre las predicciones y los valores reales
cat("RMSE: ", rmse(pred, y.test))

# Mostrar el valor del R^2 ajustado del modelo completo
cat("\nR^2 ajustado: ", summary(model.full)$adj.r.squared)

# Configurar una disposición de 2x2 para los gráficos
par(mfrow=c(2,2))

# Crear y mostrar los gráficos de diagnóstico del modelo completo
plot(model.full)
```

**Se cumplen las suposiciones de linealidad para el modelo con df_FULL.**

Podemos observar que todas las suposiciones de regresión lineal se cumplen de manera aproximada:

`Gráfico Residuals vs Fitted`

No se observa un patrón anormal en el gráfico de residuos, por lo que podemos asumir una relación lineal entre los predictores y las variables de resultado (suposición de linealidad).

`Gráfico Scale-Location`

Los residuos están distribuidos de manera uniforme en todo el rango de los predictores (suposición de homocedasticidad).

`Gráfico Normal Q-Q`

Todos los puntos se distribuyen a lo largo de la línea de referencia de manera uniforme, lo cual es suficiente para asumir que los residuos tienen una distribución normal (suposición de normalidad de los residuos).

`Gráfico Residuals vs Leverage`

Cualquier punto dentro del límite de distancia de Cook de 0.5 es influyente. Sin embargo, no hay puntos influyentes aquí (suposición de linealidad).
