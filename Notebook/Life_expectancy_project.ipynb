## 1.1 Descripción del proyecto

Este proyecto forma parte del último módulo "Programación y estadística con R" del curso de Ciencia de Datos de BEDU y se centra en identificar los determinantes clave que influyen en la esperanza de vida. La pregunta central que buscamos responder es esencial para las organizaciones de salud que buscan mejorar la esperanza de vida en regiones específicas: ¿Qué variables, y de qué manera, pueden modificarse para mejorar eficazmente la esperanza de vida?

## 1.2 Descripción del Dataset

Este dataset se generó fusionando información de dos fuentes distintas. La primera fuente proviene del Observatorio Mundial de la Salud (GHO), parte de la Organización Mundial de la Salud (OMS), que realiza un seguimiento de la salud y otros factores relacionados en 193 países. Esta información aborda aspectos de salud, incluyendo la esperanza de vida y otros indicadores relevantes. La segunda fuente corresponde al sitio web de las Naciones Unidas, proporcionando datos económicos correspondientes a los mismos países.

## 2.3 Datos atípicos

Dada la significativa proporción de datos faltantes, aproximadamente el 50%, la opción más viable es aplicar imputación. Como se explicó en la introducción, se ha observado que la falta de datos se concentra principalmente en países menos conocidos. Esto sugiere que la ausencia de estos datos podría deberse a su no disponibilidad o incluso a su inexistencia. Por lo tanto, imputar simplemente asumiendo que estos valores correspondían a ceros resulta inviable.

En cambio, una alternativa más adecuada es emplear algún valor de tendencia central. Esto reduce nuestra elección a decidir entre la media y la mediana para cada variable. Tomando en cuenta la menor sensibilidad de la mediana frente a la media con respecto a los valores atípicos, se opta por tomar esta decisión de manera individual para cada variable, considerando la cantidad de este tipo de valores en cada una. Mientras que en aquellas variables que presentan valores atípicos se elige la mediana, en aquellas sin este tipo de valores, se utiliza la media.

```r
# Crear un data frame para la gráfica
missing_data <- data.frame(
  variable = names(df),
  datos_faltantes = colSums(is.na(df))
)

# Seleccionar las variables con datos faltantes
variables_con_faltantes <- missing_data$variable[missing_data$datos_faltantes > 0]

# Función para contar outliers en una variable con base en el criterio 1.5 * IQR
count_outliers <- function(variable) {

  # Eliminar NA's
  variable <- variable[!is.na(variable)]

  # Contar outliers solo si hay valores numéricos
  if (length(variable) > 0) {
    # Calcular el rango intercuartílico (IQR)
    iqr <- IQR(variable)

    # Calcular los límites para identificar outliers
    lower_limit <- quantile(variable, 0.25) - 1.5 * iqr
    upper_limit <- quantile(variable, 0.75) + 1.5 * iqr

    # Contar outliers
    num_outliers <- sum(variable < lower_limit | variable > upper_limit)

    return(num_outliers)
  } else {
    return(NA)
  }
}

# Contar outliers solo para las variables con datos faltantes
outliers_count_per_variable <- sapply(df[variables_con_faltantes, drop = FALSE], count_outliers)

# Crear un data frame con la cantidad de outliers por variable
outliers_count_df <- data.frame(
  variable = names(outliers_count_per_variable),
  outliers_count = outliers_count_per_variable
)

# Ordenar el data frame por la cantidad de outliers de mayor a menor
outliers_count_df <- outliers_count_df[order(-outliers_count_df$outliers_count), ]

# Resetting index numbers of rows
rownames(outliers_count_df) <- NULL

# Imprimir los resultados ordenados
outliers_count_df
```

```r
# Crear un data frame para la gráfica
missing_data <- data.frame(
  variable = names(df),
  datos_faltantes = colSums(is.na(df))
)

# Seleccionar las variables con datos faltantes
variables_con_faltantes <- missing_data$variable[missing_data$datos_faltantes > 0]

# Crear la figura con 3 renglones y 5 columnas
par(mfrow = c(3, 5), mar = c(4, 4, 2, 1))

# Iterar sobre las variables y crear un boxplot para cada una
for (variable in variables_con_faltantes) {

  # Obtener el color según outliers_count_df
  color <- ifelse(outliers_count_df[outliers_count_df$variable == variable, "outliers_count"] == 0, "lightblue", "lightcoral")

  # Crear el boxplot con el color correspondiente
  boxplot(df[[variable]], main = variable, col = color, border = "black", notch = TRUE)
}

# Restablecer el diseño de la figura a uno por defecto
par(mfrow = c(1, 1), mar = c(5, 4, 4, 2) + 0.1)
```

## 2.4 Transformación de los datos

Tras haber realizado las inspecciones de datos faltantes y datos atípicos, en esta sección se manipulan los datos con el objetivo evitar problemas en los análisis posteriores.

#### Imputación de datos faltantes

Se imputa la media a los valores faltantes de las variables que no tienen outliers, en tanto que se usa la mediana para aquellas variables que sí tienen outliers.

```r
# Crear un vector para almacenar los valores de imputación
imputacion <- c()

# Iterar sobre las filas de outliers_count_df
for (i in 1:dim(outliers_count_df)[1]) {
  # Obtener el nombre de la variable actual
  variable <- outliers_count_df$variable[i]

  # Verificar si hay outliers en la variable actual
  if (outliers_count_df$outliers_count[i] > 0) {
    # Imputar con la mediana si hay outliers
    imputacion[i] <- median(df[, variable], na.rm = TRUE)
  } else {
    # Imputar con la media si no hay outliers
    imputacion[i] <- mean(df[, variable], na.rm = TRUE)
  }
}

# Aplicar la imputación al dataframe original (df)
for (i in 1:dim(outliers_count_df)[1]) {
  # Obtener el nombre de la variable actual
  variable <- outliers_count_df$variable[i]

  # Imputar los valores faltantes en df con los valores calculados
  df[, variable][is.na(df[, variable])] <- imputacion[i]
}
```

Por medio de la inspección visual percibimos un ligero sesgo hacia la izquierda (cola larga hacia la derecha). Sin embargo, para tener un valor concreto de la magnitud de esto, calculamos la asimetría de la distribución.

```r
# Asimetría de la distribución de Life_expectancy
cat("Asimetría: ", toString(skewness(df$Life_expectancy, na.rm = TRUE)))
```

Sabemos que un valor de asimetría de 0 indica una simetría perfecta. En cambio, valores entre 0 y 0.5, o entre 0 y -0.5 denotan una simetría aceptable. Por otro lado, valores por por encima de 0.5 o por debajo de -0.5 denotan asimetría significativa. Por lo tanto, en este caso tomamos medidas para mejorar la simetría de la distribución.

De manera heurística llegamos a que el escalamiento con raíz cuadrada mejora la simetría de la distribución.

```r
# Escalamiento con raiz cuadrada
df$Life_expectancy <- sqrt(max(df$Life_expectancy+1)- df$Life_expectancy)
df$Life_expectancy<- scale(df$Life_expectancy, scale=TRUE, center = TRUE)
```

```r
# Obtener la media de Life_expectancy
media_life_expectancy <- mean(df$Life_expectancy, na.rm = TRUE)

# Graficar el histograma de Life_expectancy sin grid de fondo
ggplot(df, aes(x = Life_expectancy)) +
  geom_histogram(binwidth = 1/8, fill = "lightblue", color = "black", alpha = 0.7) +
  geom_vline(xintercept = media_life_expectancy, linetype = "dashed", color = "red", size = 1) +  # Línea vertical en la media
  labs(title = "Distribución de Life_expectancy",
       x = "Años",
       y = "Frecuencia") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5, face = "bold"),
    axis.title = element_text(size = 16, margin = margin(b = 40)),
    axis.text = element_text(size = 14),
    axis.text.y = element_text(angle = 90, hjust = 0.5, margin = margin(l = 20, r = 20)),
    axis.ticks.margin = unit(0.2, "cm")
  )
```

La inspección visual denota una mejoría en la simetría de la distribución. Sin embargo, repetimos el cálculo de asimetría para corroborar.
